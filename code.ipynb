```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sc
import pandas as pd
import math
df=pd.read_csv("29_Period_of_trials_by_courts.csv")
year3=list(df.PT_1_3_Years)

year5=list(df.PT_3_5_Years)

year10=list(df.PT_5_10_Years)

year10greater=list(df.PT_Over_10_Years)
year6month=list(df.PT_Less_than_6_Months)

        
year12month=list(df.PT_6_12_Months)
year=list(set(df.Year))
total=list(df.PT_Total)

state=list(df.Area_Name)
group=set(list(df.Sub_Group_Name))
group=sorted(group)
group.pop()
#print(group)
lkar=[]
for i in range(0,len(state)):
    if(state[i]=="Karnataka"):
        lkar.append(i)
#print(year)
#print(lkar)
#print(to)
kardis5=[]
listtotalkardis=[]
for i in range(0,7):
    listtotalkardis.append(total[lkar[i]])
   # kardis5.append(year5[lkar[i]])
#print(listtotalkardis)

#print("mean of sample",meankardis)
#print(sdkardis)
meankardis=(np.mean(listtotalkardis))
sdkardis=np.std(listtotalkardis)
zkardis=(meankardis-3511)/(sdkardis)   #assuming mean to be 3511
#print(zkardis)
print("P-value of District/Session Judge is",sc.norm.cdf(zkardis)-sc.norm.cdf(-1*zkardis))
#print(sc.norm.cdf(.95))
plt.title('Box plot of District/session judge')
plt.boxplot(listtotalkardis)
plt.show()
listtotalkardis=sorted(listtotalkardis)
plt.title('Normal plot of District/session judge')
plt.plot(listtotalkardis,sc.norm.pdf(listtotalkardis,meankardis,sdkardis))
plt.show()


plt.bar(year,listtotalkardis,color="skyblue")
plt.title('District/session judge')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')
#print(listtotalkardis)

plt.axhline(meankardis,color="black")
plt.show()


listtotalkaradd=[]
for i in range(7,14):
    listtotalkaradd.append(total[lkar[i]])
#print(listtotalkaradd)
meankaradd=(np.mean(listtotalkaradd))
#print(meankaradd)
sdkaradd=np.std(listtotalkaradd)
zkaradd=(meankaradd-2680)/(sdkaradd)  #assuming mean to be 2680
#print(zkardis)
print("P-value of additional session judge is",sc.norm.cdf(zkaradd)-sc.norm.cdf(-1*zkaradd)) 
plt.title('Box plot of Additional session judge')
plt.boxplot(listtotalkaradd)
plt.show()
listtotalkaradd=sorted(listtotalkaradd)
plt.title('Normal plot of Additional session judge')
plt.plot(listtotalkaradd,sc.norm.pdf(listtotalkaradd,meankaradd,sdkaradd))
plt.show()
plt.bar(year,listtotalkaradd,color="pink")
plt.title('Additional session judge')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')
meankaradd=(np.mean(listtotalkaradd))
plt.axhline(meankaradd,color="black")
plt.show()


listtotalkarchief=[]
for i in range(14,21):
    listtotalkarchief.append(total[lkar[i]])
#print(listtotalkarchief)
plt.title('box plot of Chief Judicial magistrate')
plt.boxplot(listtotalkarchief)
plt.show()
plt.bar(year,listtotalkarchief,color="lightgreen")
plt.title('Chief Judicial magistrate')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')

meankarchief=(np.mean(listtotalkarchief))
plt.axhline(meankarchief,color="black")
plt.show()
plt.title('No of cases pending in Chief Judicial Magistrate in the year 2006')
lpiechief=[5916,3707,1561,2964,1375,111]
s=["1-3 Years","3-5 Years","5-10 Years","6-12 Months","Less than 6 months","More than 10 Years"]
plt.pie(lpiechief,labels=s,autopct='%.1f%%')
plt.show()

listtotalkarjm1=[]
for i in range(21,28):
    listtotalkarjm1.append(total[lkar[i]])
#print(listtotalkarjm1)
plt.boxplot(listtotalkarjm1)
plt.show()
plt.bar(year,listtotalkarjm1,color="orange")
plt.title('Judicial Magistrate 1')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')
meankarjm1=(np.mean(listtotalkarjm1))
plt.axhline(meankarjm1,color="black")
plt.show()

plt.title('No of cases pending in  Judicial Magistrate-I in the year 2004')
lpiechief=[7531,3931,1608,4896,4689,201]
s=["1-3 Years","3-5 Years","5-10 Years","6-12 Months","Less than 6 months","More than 10 Years"]
plt.pie(lpiechief,labels=s,autopct='%.1f%%')
plt.show()

plt.title('No of cases pending in Judicial Magistrate-I in the year 2008')
lpiechief=[11314,5037,1772,2802,8181,226]
s=["1-3 Years","3-5 Years","5-10 Years","6-12 Months","Less than 6 months","More than 10 Years"]
plt.pie(lpiechief,labels=s,autopct='%.1f%%')
plt.show()


listtotalkarjm2=[]
for i in range(28,35):
    listtotalkarjm2.append(total[lkar[i]])
#print(listtotalkarjm2)
plt.title('Box plot of Judicial Magistrate 2')
plt.boxplot(listtotalkarjm2)
plt.show()
plt.bar(year,listtotalkarjm2,color="yellow")
plt.title('Judicial Magistrate 2')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')
meankarjm2=(np.mean(listtotalkarjm2))
plt.axhline(meankarjm2,color="black")
plt.show()
plt.title('No of cases pending in Judicial Magistrate-II in the year 2004')
lpiechief=[5089,2714,832,4626,3999,128]
s=["1-3 Years","3-5 Years","5-10 Years","6-12 Months","Less than 6 months","More than 10 Years"]
plt.pie(lpiechief,labels=s,autopct='%.1f%%')
plt.show()



listtotalkarsjm=[]
for i in range(35,42):
    listtotalkarsjm.append(total[lkar[i]])
#print(listtotalkarsjm)
plt.title('Box plot of Special Judicial Magistrate')
plt.boxplot(listtotalkarsjm)
plt.show()
plt.bar(year,listtotalkarsjm,color="grey")
plt.title('Special Judicial Magistrate')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')
meankarsjm=(np.mean(listtotalkarsjm))
plt.axhline(meankarsjm,color="black")
plt.show()
plt.title('No of cases pending in Special Judicial Magistrate in the year 2006')
lpie=[679,562,201,617,202,0]
s=["1-3 Years","3-5 Years","5-10 Years","6-12 Months","Less than 6 months","More than 10 Years"]
plt.pie(lpie,labels=s,autopct='%.1f%%')
plt.show()


listtotalkaroc=[]
for i in range(42,49):
    listtotalkaroc.append(total[lkar[i]])
#print(listtotalkaroc)
plt.boxplot(listtotalkaroc)
plt.show()
plt.bar(year,listtotalkaroc,color="brown")
plt.title('Other Courts')
#plt.set_xlabel('Years')
plt.ylabel('Total_no_of_cases_pending')
plt.xlabel('Years')
meankaroc=(np.mean(listtotalkaroc))
plt.axhline(meankaroc,color="black")
plt.show()

```
